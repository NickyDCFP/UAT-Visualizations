```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_sign: np.ndarray = np.array([0, 1, 1, 6]) - np.pi
f_sign: np.ndarray = np.array([0, 0, 5, 5])
x: np.ndarray = np.linspace(1, 2 * np.pi, 300) - np.pi
f: np.ndarray = np.sin(x - 1) + 5
```

```python
plt.plot(x, f, label="True Function")
plt.plot(x_sign, f_sign, label="Single Neuron")
plt.title("Single Neuron Left-Side Approximation")
plt.legend()
plt.show()
```

```python
x_sign: np.ndarray = np.array([0, 1, 1, 1.5, 1.5, 6]) - np.pi
f_sign: np.ndarray = np.array([0, 0, 5, 5, np.sin(0.5 - np.pi) + 5, np.sin(0.5 - np.pi) + 5])
x: np.ndarray = np.linspace(1, 2 * np.pi, 300) - np.pi
f: np.ndarray = np.sin(x - 1) + 5
```

```python
plt.plot(x, f, label="True Function")
plt.plot(x_sign, f_sign, label="Two Neurons")
plt.title("Two Neuron Left-Side Approximation")
plt.legend()
plt.show()
```

```python
x: np.ndarray = np.linspace(-np.pi, np.pi, 300)
f: np.ndarray = np.sin(x)
```

```python
x_approx: np.ndarray = np.linspace(-np.pi, np.pi, 1200)
f_approx: np.ndarray = [np.sin(x_approx[i - i % 25]) for i in range(len(x_approx))]
```

```python
plt.plot(x, f, label="True Function")
plt.plot(x_approx, f_approx, label="UAT-Approximated")
plt.axis('off')
plt.title("Universal Approximation Theorem Intuition")
plt.legend()
plt.show()
```

```python
x1: np.ndarray = np.linspace(-np.pi, np.pi, 300)
x2: np.ndarray = np.linspace(-np.pi, np.pi, 300)
X1: np.ndarray; X2: np.ndarray
X1, X2 = np.meshgrid(x1, x2)
f: np.ndarray = np.sin(X1) + np.cos(X2)
```

```python
x1_approx: np.ndarray = np.linspace(-np.pi, np.pi, 1200)
x2_approx: np.ndarray = np.linspace(-np.pi, np.pi, 1200)
x1_repeat: np.ndarray = np.array(x1_approx)
x2_repeat: np.ndarray = np.array(x2_approx)
for i in range(len(x1_repeat)):
    x1_repeat[i] = x1_repeat[i - i % 50]
    x2_repeat[i] = x2_repeat[i - i % 50]
X1_approx: np.ndarray; X2_approx: np.ndarray
X1_APPROX, X2_APPROX = np.meshgrid(x1_approx, x2_approx)
X1_REPEAT, X2_REPEAT = np.meshgrid(x1_repeat, x2_repeat)
f_approx: np.ndarray = np.sin(X1_REPEAT) + np.cos(X2_REPEAT)
```

```python
from matplotlib.figure import Figure
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import colormaps
fig: Figure = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(6)
fig.suptitle("High-Dimensional UAT Intuition", size='xx-large')
ax_true: Axes3D = fig.add_subplot(131, projection='3d')
ax_approx: Axes3D = fig.add_subplot(132, projection='3d')
ax_both: Axes3D = fig.add_subplot(133, projection='3d')
ax_true.set(
       xticklabels=[],
       yticklabels=[],
       zticklabels=[],
       title='True Function'
)
ax_approx.set(
       xticklabels=[],
       yticklabels=[],
       zticklabels=[],
       title='Approximated Function'
)
ax_both.set(
       xticklabels=[],
       yticklabels=[],
       zticklabels=[],
       title='True and Approximated Functions Overlaid'
)

surface_true_independent = ax_true.plot_surface(
    X1,
    X2,
    f,
    cmap=colormaps['terrain'],
    alpha=1
)
surface_approx_independent = ax_approx.plot_surface(
    X1_APPROX,
    X2_APPROX,
    f_approx,
    cmap=colormaps['terrain'],
    alpha=1
)
surface_approx_overlaid = ax_both.plot_surface(
    X1_APPROX,
    X2_APPROX,
    f_approx,
    cmap=colormaps['gray'],
    alpha=0.2
)
surface_true_overlaid = ax_both.plot_surface(
    X1,
    X2,
    f,
    cmap=colormaps['terrain'],
    alpha=0.8
)
plt.show()
```
